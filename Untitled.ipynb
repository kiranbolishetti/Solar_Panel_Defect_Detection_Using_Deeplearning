```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Lambda # Imported Lambda
import os
import matplotlib.pyplot as plt

# --- CONFIGURATION ---
# IMPORTANT: Change this to the path where your dataset folders are located
DATASET_PATH = 'Solar_Panel_Dataset' 
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
NUM_CLASSES = 6
MODEL_SAVE_PATH = 'solar_panel_classifier.h5'
EPOCHS = 20 # Start with 20 epochs, you may need more if accuracy is low

# --- 1. Data Loading and Preprocessing ---
print("Setting up Data Generators...")

# The generator will only handle augmentation and batching. 
# Normalization (rescale=1./255) is applied here, but will be overridden 
# by the MobileNetV2 preprocessing function inside the model.
datagen = ImageDataGenerator(
    rescale=1./255, 
    validation_split=0.2, 
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Load data from directories
train_generator = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    seed=42
)

validation_generator = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    seed=42
)

print("\nClassification Class Indices (Alphabetical):", train_generator.class_indices)

# --- 2. Build the Transfer Learning Model (MobileNetV2) ---
print("\nBuilding MobileNetV2 Transfer Learning Model...")

# 1. Load the pre-trained MobileNetV2 model, excluding its final classification layer
base_model = MobileNetV2(
    input_shape=IMAGE_SIZE + (3,),
    include_top=False,
    weights='imagenet' 
)

# 2. Freeze the weights of the pre-trained layers
base_model.trainable = False

# 3. Create the new classification head for our 6 classes
model = Sequential([
    # FIX: Wrap the preprocessing function in a Lambda layer to use it in Sequential model
    Lambda(
        tf.keras.applications.mobilenet_v2.preprocess_input, 
        input_shape=IMAGE_SIZE + (3,)
    ),
    
    base_model,
    GlobalAveragePooling2D(), 
    Dense(NUM_CLASSES, activation='softmax') 
])

# --- 3. Compile and Train the Model ---

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), 
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("\nStarting Model Training (Transfer Learning Feature Extraction)...")

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // BATCH_SIZE
)

# --- 4. Save the Final Model ---
model.save(MODEL_SAVE_PATH)
print(f"\n✅ Model training complete and saved successfully as '{MODEL_SAVE_PATH}'")

# Optional: Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```
