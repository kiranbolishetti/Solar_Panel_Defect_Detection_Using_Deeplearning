```python
"""
Utility functions for SolarGuard project
"""
import os
import shutil
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import cv2
from pathlib import Path
import json

def create_project_structure():
    """Create necessary directories for the project"""
    directories = [
        'data/raw',
        'data/processed/train',
        'data/processed/val',
        'data/processed/test',
        'models',
        'results',
        'visualizations'
    ]
    
    for directory in directories:
        Path(directory).mkdir(parents=True, exist_ok=True)
    print("‚úÖ Project structure created successfully!")

def load_image(image_path, target_size=(224, 224)):
    """Load and preprocess a single image"""
    try:
        img = Image.open(image_path).convert('RGB')
        img = img.resize(target_size)
        return np.array(img)
    except Exception as e:
        print(f"Error loading image {image_path}: {e}")
        return None

def plot_images_grid(images, labels, predictions=None, n_cols=5, figsize=(15, 10)):
    """Plot a grid of images with labels"""
    n_images = len(images)
    n_rows = (n_images + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
    axes = axes.flatten() if n_images > 1 else [axes]
    
    for idx, ax in enumerate(axes):
        if idx < n_images:
            ax.imshow(images[idx])
            title = f"True: {labels[idx]}"
            if predictions is not None:
                title += f"\nPred: {predictions[idx]}"
            ax.set_title(title, fontsize=10)
            ax.axis('off')
        else:
            ax.axis('off')
    
    plt.tight_layout()
    return fig

def calculate_class_weights(label_counts):
    """Calculate class weights for imbalanced datasets"""
    total = sum(label_counts.values())
    weights = {cls: total / (len(label_counts) * count) 
               for cls, count in label_counts.items()}
    return weights

def save_model_history(history, filepath):
    """Save training history to JSON"""
    hist_dict = {key: [float(val) for val in values] 
                 for key, values in history.history.items()}
    with open(filepath, 'w') as f:
        json.dump(hist_dict, f, indent=4)
    print(f"‚úÖ History saved to {filepath}")

def plot_training_history(history, save_path=None):
    """Plot training and validation metrics"""
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Accuracy
    axes[0].plot(history.history['accuracy'], label='Train Accuracy')
    axes[0].plot(history.history['val_accuracy'], label='Val Accuracy')
    axes[0].set_title('Model Accuracy', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Accuracy')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    
    # Loss
    axes[1].plot(history.history['loss'], label='Train Loss')
    axes[1].plot(history.history['val_loss'], label='Val Loss')
    axes[1].set_title('Model Loss', fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Loss')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"‚úÖ Training history saved to {save_path}")
    
    return fig

def normalize_image(image):
    """Normalize image to [0, 1] range"""
    return image.astype(np.float32) / 255.0

def denormalize_image(image):
    """Convert normalized image back to [0, 255] range"""
    return (image * 255).astype(np.uint8)

def get_class_distribution(data_dir):
    """Get distribution of classes in dataset"""
    class_counts = {}
    for class_name in os.listdir(data_dir):
        class_path = os.path.join(data_dir, class_name)
        if os.path.isdir(class_path):
            count = len([f for f in os.listdir(class_path) 
                        if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
            class_counts[class_name] = count
    return class_counts
```

```python
import os
import shutil
import numpy as np
from pathlib import Path
from sklearn.model_selection import train_test_split
import albumentations as A
from PIL import Image
import cv2
from tqdm import tqdm

class SolarPanelDataProcessor:
    """Data preprocessing and augmentation for solar panel images"""
    
    def __init__(self, raw_data_path, processed_data_path, img_size=224):
        self.raw_data_path = Path(raw_data_path)
        self.processed_data_path = Path(processed_data_path)
        self.img_size = img_size
        self.classes = ['Clean', 'Dusty', 'Bird-Drop', 'Electrical-Damage', 
                       'Physical-Damage', 'Snow-Covered']
        
    def get_augmentation_pipeline(self, is_train=True):
        """Define augmentation pipeline"""
        if is_train:
            return A.Compose([
                A.Resize(self.img_size, self.img_size),
                A.HorizontalFlip(p=0.5),
                A.VerticalFlip(p=0.3),
                A.RandomRotate90(p=0.5),
                A.RandomBrightnessContrast(p=0.5),
            ])
        else:
            return A.Compose([
                A.Resize(self.img_size, self.img_size)
            ])
    
    def organize_data(self, train_split=0.7, val_split=0.15, test_split=0.15):
        """Organize raw data into train/val/test splits"""
        print("üìÇ Organizing dataset...")
        
        for split in ['train', 'val', 'test']:
            for cls in self.classes:
                (self.processed_data_path / split / cls).mkdir(parents=True, exist_ok=True)
        
        for cls in self.classes:
            class_path = self.raw_data_path / cls
            if not class_path.exists():
                print(f"‚ö†Ô∏è  {cls} directory not found - SKIPPING")
                continue
            
            images = list(class_path.glob('*.jpg')) + list(class_path.glob('*.jpeg')) + list(class_path.glob('*.png'))
            
            if len(images) == 0:
                print(f"‚ö†Ô∏è  {cls}: No images - SKIPPING")
                continue
            
            train_imgs, temp_imgs = train_test_split(images, test_size=(1-train_split), random_state=42)
            val_imgs, test_imgs = train_test_split(temp_imgs, test_size=test_split/(val_split+test_split), random_state=42)
            
            for img_path in train_imgs:
                shutil.copy2(img_path, self.processed_data_path / 'train' / cls / img_path.name)
            for img_path in val_imgs:
                shutil.copy2(img_path, self.processed_data_path / 'val' / cls / img_path.name)
            for img_path in test_imgs:
                shutil.copy2(img_path, self.processed_data_path / 'test' / cls / img_path.name)
            
            print(f"‚úÖ {cls}: Train={len(train_imgs)}, Val={len(val_imgs)}, Test={len(test_imgs)}")
        
        print("\n‚úÖ Data organization complete!")
    
    def augment_minority_classes(self, target_count=500):
        """Augment minority classes - FIXED VERSION"""
        print("\nüîÑ Augmenting minority classes...")
        transform = self.get_augmentation_pipeline(is_train=True)
        
        train_path = self.processed_data_path / 'train'
        
        for cls in self.classes:
            class_path = train_path / cls
            
            if not class_path.exists():
                print(f"‚ö†Ô∏è  {cls}: Not found - SKIPPING")
                continue
            
            images = list(class_path.glob('*.jpg')) + list(class_path.glob('*.jpeg')) + list(class_path.glob('*.png'))
            current_count = len(images)
            
            # FIXED: Check if empty BEFORE trying to augment
            if current_count == 0:
                print(f"‚ö†Ô∏è  {cls}: EMPTY - SKIPPING")
                continue
            
            if current_count < target_count:
                needed = target_count - current_count
                print(f"üìà Augmenting {cls}: {current_count} -> {target_count} (+{needed})")
                
                for i in tqdm(range(needed), desc=f"Augmenting {cls}"):
                    img_path = np.random.choice(images)
                    img = cv2.imread(str(img_path))
                    if img is None:
                        continue
                    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                    augmented = transform(image=img)
                    aug_img = augmented['image']
                    save_name = f"{img_path.stem}_aug_{i}.jpg"
                    save_path = class_path / save_name
                    aug_img_bgr = cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR)
                    cv2.imwrite(str(save_path), aug_img_bgr)
            else:
                print(f"‚úÖ {cls}: {current_count} images (no augmentation needed)")
        
        print("\n‚úÖ Augmentation complete!")

# NOW RUN THIS:
processor = SolarPanelDataProcessor(
    raw_data_path=r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard\Solar_Panel_Dataset",
    processed_data_path=r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard\processed_data",
    img_size=224
)


print("üîç Checking your dataset:")
for cls in processor.classes:
    cls_path = processor.raw_data_path / cls
    images = (
        list(cls_path.glob("*.jpg")) +
        list(cls_path.glob("*.jpeg")) +
        list(cls_path.glob("*.png"))
    ) if cls_path.exists() else []
    
    print(f"{cls}: {len(images)} images")

processor.organize_data()
processor.augment_minority_classes(target_count=500)
```

```python
"""
Exploratory Data Analysis for Solar Panel Detection
"""
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from PIL import Image
import cv2

class SolarPanelEDA:
    """EDA class for solar panel dataset"""
    
    def __init__(self, data_path):
        self.data_path = Path(data_path)
        self.classes = ['Clean', 'Dusty', 'Bird-Drop', 'Electrical-Damage', 
                        'Physical-Damage', 'Snow-Covered']
        self.colors = ['#2ecc71', '#f39c12', '#8e44ad', '#e74c3c', '#34495e', '#3498db']
        
        # Set style
        sns.set_style("whitegrid")
        plt.rcParams['figure.figsize'] = (15, 10)
    
    def get_class_distribution(self):
        """Get distribution of classes across splits"""
        data = []
        
        for split in ['train', 'val', 'test']:
            split_path = self.data_path / split
            if not split_path.exists():
                continue
            
            for cls in self.classes:
                class_path = split_path / cls
                if class_path.exists():
                    count = len(list(class_path.glob('*.jpg')) + 
                                list(class_path.glob('*.jpeg')) + 
                                list(class_path.glob('*.png')))
                    data.append({
                        'Split': split,
                        'Class': cls,
                        'Count': count
                    })
                else:
                    data.append({'Split': split, 'Class': cls, 'Count': 0})
        
        return pd.DataFrame(data)
    
    def plot_class_distribution(self, save_path=None):
        """Plot class distribution across splits"""
        df = self.get_class_distribution()
        
        fig, axes = plt.subplots(2, 2, figsize=(18, 12))
        
        # 1. Overall distribution
        overall = df.groupby('Class')['Count'].sum().sort_values(ascending=False)
        axes[0, 0].bar(range(len(overall)), overall.values, color=self.colors[:len(overall)])
        axes[0, 0].set_xticks(range(len(overall)))
        axes[0, 0].set_xticklabels(overall.index, rotation=45, ha='right')
        axes[0, 0].set_title('Overall Class Distribution', fontsize=14, fontweight='bold')
        axes[0, 0].set_ylabel('Count')
        axes[0, 0].grid(axis='y', alpha=0.3)
        for i, v in enumerate(overall.values):
            axes[0, 0].text(i, v + 10, str(v), ha='center', fontweight='bold')
        
        # 2. Distribution by split
        pivot_df = df.pivot(index='Class', columns='Split', values='Count').fillna(0)
        pivot_df.plot(kind='bar', ax=axes[0, 1], color=['#3498db', '#e74c3c', '#2ecc71'])
        axes[0, 1].set_title('Class Distribution by Split', fontsize=14, fontweight='bold')
        axes[0, 1].set_ylabel('Count')
        axes[0, 1].set_xlabel('Class')
        axes[0, 1].legend(title='Split')
        axes[0, 1].tick_params(axis='x', rotation=45)
        axes[0, 1].grid(axis='y', alpha=0.3)
        
        # 3. Pie chart (training set)
        train_df = df[df['Split'] == 'train']
        if len(train_df) > 0 and train_df['Count'].sum() > 0:
            axes[1, 0].pie(train_df['Count'], labels=train_df['Class'], autopct='%1.1f%%',
                          colors=self.colors[:len(train_df)], startangle=90)
            axes[1, 0].set_title('Training Set Distribution', fontsize=14, fontweight='bold')
        else:
            axes[1, 0].text(0.5, 0.5, "No training images found", ha='center', va='center')
        
        # 4. Stacked bar chart
        if not pivot_df.empty:
            pivot_df.T.plot(kind='bar', stacked=True, ax=axes[1, 1], 
                           color=self.colors[:len(self.classes)])
            axes[1, 1].set_title('Split Distribution by Class', fontsize=14, fontweight='bold')
            axes[1, 1].set_ylabel('Count')
            axes[1, 1].set_xlabel('Split')
            axes[1, 1].legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
            axes[1, 1].tick_params(axis='x', rotation=0)
            axes[1, 1].grid(axis='y', alpha=0.3)
        
        plt.tight_layout()
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"‚úÖ Class distribution plot saved to {save_path}")
        plt.show()
        return fig
    
    def generate_eda_report(self):
        """Generate comprehensive EDA report"""
        print("\n" + "="*70)
        print("üìä SOLAR PANEL DATASET - EDA REPORT")
        print("="*70)
        
        df = self.get_class_distribution()
        if df['Count'].sum() == 0:
            print("‚ùå No images found. Check your `data_path`!")
            return
        
        # Class distribution
        print("\n1Ô∏è‚É£  CLASS DISTRIBUTION:")
        print("-" * 70)
        for split in ['train', 'val', 'test']:
            split_df = df[df['Split'] == split]
            if len(split_df) > 0:
                print(f"\n{split.upper()}:")
                for _, row in split_df.iterrows():
                    print(f"  {row['Class']:20s}: {row['Count']:4d} images")
                print(f"  {'TOTAL':20s}: {split_df['Count'].sum():4d} images")
        
        # Overall stats
        print("\n2Ô∏è‚É£  OVERALL STATISTICS:")
        print("-" * 70)
        total_images = df['Count'].sum()
        print(f"Total images: {total_images}")
        print(f"Number of classes: {len(self.classes)}")
        
        # Class imbalance (safe division)
        class_totals = df.groupby('Class')['Count'].sum()
        max_class = class_totals.idxmax()
        min_class = class_totals.idxmin()
        if class_totals.min() > 0:
            imbalance_ratio = class_totals.max() / class_totals.min()
            print(f"\nMost common class: {max_class} ({class_totals.max()} images)")
            print(f"Least common class: {min_class} ({class_totals.min()} images)")
            print(f"Imbalance ratio: {imbalance_ratio:.2f}:1")
        else:
            print("‚ùå Cannot calculate imbalance ratio - some classes have 0 images")
        print("\n" + "="*70)


# ------------------- USAGE -------------------

if __name__ == "__main__":
    # Initialize EDA with your correct processed dataset path
    eda = SolarPanelEDA(
        data_path=r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard\processed_data"
    )
    
    # Generate EDA report
    eda.generate_eda_report()
    
    # Create visualizations folder
    os.makedirs('visualizations', exist_ok=True)
    
    # Plots
    eda.plot_class_distribution(save_path='visualizations/class_distribution.png')
```

```python
import os
import shutil
import random

# Original dataset
src = r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard\Solar_Panel_Dataset"

# Destination folder with splits
dst = r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard\Solar_Panel_Dataset_split"
os.makedirs(dst, exist_ok=True)

# Splits
splits = ['train', 'val', 'test']
for split in splits:
    os.makedirs(os.path.join(dst, split), exist_ok=True)

# Go through each class and split
for cls in os.listdir(src):
    cls_path = os.path.join(src, cls)
    if not os.path.isdir(cls_path):
        continue
    
    # Get only image files
    images = [f for f in os.listdir(cls_path) 
              if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    
    if len(images) == 0:
        print(f"‚ö†Ô∏è No images found in class folder: {cls}")
        continue

    random.shuffle(images)
    
    n_total = len(images)
    n_train = int(0.7 * n_total)
    n_val = int(0.15 * n_total)
    n_test = n_total - n_train - n_val
    
    idx = 0
    for split, count in zip(splits, [n_train, n_val, n_test]):
        split_cls_path = os.path.join(dst, split, cls)
        os.makedirs(split_cls_path, exist_ok=True)
        for img in images[idx:idx+count]:
            shutil.copy(os.path.join(cls_path, img), split_cls_path)
        idx += count

print("‚úÖ Dataset split into train/val/test successfully!")
```

```python
import os
import shutil
import random

# PATHS (CHANGE NOTHING)
SRC = r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard2\Solar_Panel_Dataset"
DST = r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard2\data\processed"

classes = [
    "Clean",
    "Dusty",
    "Bird-Drop",
    "Electrical-Damage",
    "Physical-Damage",
    "Snow-Covered"
]

# Split ratios
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

# Create folders
for split in ['train', 'val', 'test']:
    for cls in classes:
        os.makedirs(os.path.join(DST, split, cls), exist_ok=True)

# Split and copy images
for cls in classes:
    cls_src = os.path.join(SRC, cls)
    images = [f for f in os.listdir(cls_src) if f.lower().endswith(('.jpg','.jpeg','.png'))]
    random.shuffle(images)

    n = len(images)
    train_end = int(n * train_ratio)
    val_end = train_end + int(n * val_ratio)

    splits = {
        'train': images[:train_end],
        'val': images[train_end:val_end],
        'test': images[val_end:]
    }

    for split, imgs in splits.items():
        for img in imgs:
            src_img = os.path.join(cls_src, img)
            dst_img = os.path.join(DST, split, cls, img)
            shutil.copy2(src_img, dst_img)

    print(f"{cls}: {len(images)} images split")

print("\n‚úÖ Dataset successfully split into train / val / test")
```

```python
from PIL import Image
import os

BASE_DIR = r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard2\data\processed\train"
TARGET_SIZE = (224, 224)

classes = ['Clean', 'Dusty', 'Bird-Drop']

for cls in classes:
    class_dir = os.path.join(BASE_DIR, cls)
    
    for img_name in os.listdir(class_dir):
        img_path = os.path.join(class_dir, img_name)
        
        try:
            img = Image.open(img_path).convert("RGB")
            img = img.resize(TARGET_SIZE)
            img.save(img_path)
        except Exception as e:
            print(f"‚ùå Error resizing {img_path}: {e}")

print("‚úÖ All training images resized to 224x224")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 10

DATA_DIR = r"C:\Users\Kiran Kumar\Desktop\AllProejects\Project_5\SolarGuard\data\processed"

# Data generators
datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.2,
    horizontal_flip=True
)

train_gen = datagen.flow_from_directory(
    DATA_DIR + r"\train",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical"
)

val_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    DATA_DIR + r"\val",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical"
)

test_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    DATA_DIR + r"\test",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    shuffle=False
)

# Model
base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(224,224,3))
base_model.trainable = False

x = GlobalAveragePooling2D()(base_model.output)
x = Dense(128, activation="relu")(x)
x = Dropout(0.5)(x)
output = Dense(train_gen.num_classes, activation="softmax")(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# Train
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS
)

print("‚úÖ Training completed")
```

```python
from sklearn.metrics import classification_report
import numpy as np

preds = model.predict(test_gen)
y_pred = np.argmax(preds, axis=1)

print("\nüìã CLASSIFICATION REPORT\n")
print(classification_report(test_gen.classes, y_pred, target_names=list(test_gen.class_indices.keys())))
```

```python
"""
YOLOv8 Object Detection Training for Solar Panel Defects
Fixed matplotlib import issue
"""
import os
from pathlib import Path
from ultralytics import YOLO
import yaml
import cv2
import numpy as np

# Fix matplotlib import issue
try:
    import matplotlib
    matplotlib.use('Agg')  # Use non-interactive backend
    import matplotlib.pyplot as plt
    from PIL import Image
    MATPLOTLIB_AVAILABLE = True
except (ImportError, AttributeError) as e:
    print(f"‚ö†Ô∏è  Matplotlib not available: {e}")
    print("   Visualization features will be disabled")
    MATPLOTLIB_AVAILABLE = False
    try:
        from PIL import Image
    except ImportError:
        Image = None

class SolarPanelObjectDetector:
    """Object Detection for Solar Panel Defects using YOLOv8"""
    
    def __init__(self, data_yaml_path, model_size='n'):
        """
        Initialize YOLO detector
        
        Args:
            data_yaml_path: Path to data.yaml configuration
            model_size: 'n' (nano), 's' (small), 'm' (medium), 'l' (large), 'x' (xlarge)
        """
        self.data_yaml_path = data_yaml_path
        self.model_size = model_size
        self.model = None
        self.results = None
        
    def create_data_yaml(self, dataset_path, class_names):
        """
        Create YOLO data.yaml configuration file
        
        Args:
            dataset_path: Path to dataset root
            class_names: List of class names
        """
        data_config = {
            'path': str(Path(dataset_path).absolute()),
            'train': 'images/train',
            'val': 'images/val',
            'test': 'images/test',
            'nc': len(class_names),
            'names': class_names
        }
        
        # Create parent directory if it doesn't exist
        yaml_path = Path(self.data_yaml_path)
        yaml_path.parent.mkdir(parents=True, exist_ok=True)
        
        with open(self.data_yaml_path, 'w') as f:
            yaml.dump(data_config, f, default_flow_style=False)
        
        print(f"‚úÖ Created data.yaml at {self.data_yaml_path}")
        return data_config
    
    def prepare_yolo_dataset(self, source_dir, output_dir):
        """
        Prepare dataset in YOLO format
        Expects annotations in YOLO format: class x_center y_center width height
        """
        print("\nüì¶ Preparing YOLO dataset...")
        
        splits = ['train', 'val', 'test']
        
        for split in splits:
            # Create directories
            img_dir = Path(output_dir) / 'images' / split
            label_dir = Path(output_dir) / 'labels' / split
            img_dir.mkdir(parents=True, exist_ok=True)
            label_dir.mkdir(parents=True, exist_ok=True)
            
            print(f"‚úÖ Created directories for {split} split")
        
        print("‚úÖ YOLO dataset structure ready!")
        print("\n‚ö†Ô∏è  Note: You need to annotate images using tools like:")
        print("   - LabelImg: https://github.com/HumanSignal/labelImg")
        print("   - Roboflow: https://roboflow.com/annotate")
        print("   Save annotations in YOLO format (.txt files)")
    
    def load_model(self, pretrained=True):
        """Load YOLOv8 model"""
        if pretrained:
            model_name = f'yolov8{self.model_size}.pt'
            print(f"\nüî® Loading pretrained YOLOv8{self.model_size.upper()} model...")
        else:
            model_name = f'yolov8{self.model_size}.yaml'
            print(f"\nüî® Creating YOLOv8{self.model_size.upper()} model from scratch...")
        
        self.model = YOLO(model_name)
        print("‚úÖ Model loaded successfully!")
        return self.model
    
    def train(self, epochs=100, imgsz=640, batch=16, patience=50):
        """
        Train YOLOv8 model
        
        Args:
            epochs: Number of training epochs
            imgsz: Input image size
            batch: Batch size
            patience: Early stopping patience
        """
        print("\nüöÄ Starting YOLOv8 training...")
        
        self.results = self.model.train(
            data=self.data_yaml_path,
            epochs=epochs,
            imgsz=imgsz,
            batch=batch,
            patience=patience,
            save=True,
            project='runs/detect',
            name='solar_panel_yolo',
            exist_ok=True,
            pretrained=True,
            optimizer='AdamW',
            lr0=0.01,
            lrf=0.01,
            momentum=0.937,
            weight_decay=0.0005,
            warmup_epochs=3,
            warmup_momentum=0.8,
            box=7.5,
            cls=0.5,
            dfl=1.5,
            hsv_h=0.015,
            hsv_s=0.7,
            hsv_v=0.4,
            degrees=0.0,
            translate=0.1,
            scale=0.5,
            shear=0.0,
            perspective=0.0,
            flipud=0.0,
            fliplr=0.5,
            mosaic=1.0,
            mixup=0.0,
            copy_paste=0.0,
            verbose=True
        )
        
        print("\n‚úÖ Training complete!")
        print(f"üìä Results saved to: {self.results.save_dir}")
        
        return self.results
    
    def validate(self):
        """Validate the model"""
        print("\nüìä Validating model...")
        metrics = self.model.val()
        
        print("\nüìã Validation Metrics:")
        print(f"  mAP50: {metrics.box.map50:.4f}")
        print(f"  mAP50-95: {metrics.box.map:.4f}")
        print(f"  Precision: {metrics.box.mp:.4f}")
        print(f"  Recall: {metrics.box.mr:.4f}")
        
        return metrics
    
    def predict(self, image_path, conf=0.25, save=True):
        """
        Make prediction on an image
        
        Args:
            image_path: Path to image
            conf: Confidence threshold
            save: Save annotated image
        """
        results = self.model.predict(
            source=image_path,
            conf=conf,
            save=save,
            project='runs/detect',
            name='predictions',
            exist_ok=True
        )
        
        return results
    
    def predict_batch(self, images_dir, conf=0.25):
        """Predict on multiple images"""
        results = self.model.predict(
            source=images_dir,
            conf=conf,
            save=True,
            project='runs/detect',
            name='batch_predictions',
            exist_ok=True
        )
        
        return results
    
    def visualize_predictions(self, image_path, conf=0.25, save_path=None):
        """Visualize predictions on an image"""
        if not MATPLOTLIB_AVAILABLE:
            print("‚ö†Ô∏è  Matplotlib not available. Using YOLO's built-in visualization.")
            return self.predict(image_path, conf=conf, save=True)
        
        results = self.predict(image_path, conf=conf, save=False)
        
        # Get original image
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        
        # Plot
        fig, axes = plt.subplots(1, 2, figsize=(16, 8))
        
        # Original
        axes[0].imshow(img)
        axes[0].set_title('Original Image', fontsize=14, fontweight='bold')
        axes[0].axis('off')
        
        # Predictions
        result_img = results[0].plot()
        axes[1].imshow(result_img)
        axes[1].set_title(f'Detections (conf > {conf})', fontsize=14, fontweight='bold')
        axes[1].axis('off')
        
        plt.tight_layout()
        
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"‚úÖ Visualization saved to {save_path}")
        
        plt.show()
        return fig
    
    def export_model(self, format='onnx'):
        """
        Export model to different formats
        
        Args:
            format: 'onnx', 'torchscript', 'coreml', 'tflite', etc.
        """
        print(f"\nüì¶ Exporting model to {format.upper()}...")
        path = self.model.export(format=format)
        print(f"‚úÖ Model exported to: {path}")
        return path
    
    def plot_training_results(self, results_dir='runs/detect/solar_panel_yolo'):
        """Plot training results"""
        if not MATPLOTLIB_AVAILABLE or Image is None:
            print("‚ö†Ô∏è  Matplotlib/PIL not available for plotting.")
            print(f"   Check results manually at: {results_dir}")
            return
        
        results_path = Path(results_dir)
        
        # Check if results exist
        if not results_path.exists():
            print(f"‚ö†Ô∏è  Results directory not found: {results_dir}")
            return
        
        # Plot results
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        
        # Load and plot results.png if available
        results_img_path = results_path / 'results.png'
        if results_img_path.exists():
            img = Image.open(results_img_path)
            axes[0, 0].imshow(img)
            axes[0, 0].set_title('Training Results', fontsize=14, fontweight='bold')
            axes[0, 0].axis('off')
        
        # Confusion matrix
        cm_path = results_path / 'confusion_matrix.png'
        if cm_path.exists():
            img = Image.open(cm_path)
            axes[0, 1].imshow(img)
            axes[0, 1].set_title('Confusion Matrix', fontsize=14, fontweight='bold')
            axes[0, 1].axis('off')
        
        # PR curve
        pr_path = results_path / 'PR_curve.png'
        if pr_path.exists():
            img = Image.open(pr_path)
            axes[1, 0].imshow(img)
            axes[1, 0].set_title('Precision-Recall Curve', fontsize=14, fontweight='bold')
            axes[1, 0].axis('off')
        
        # F1 curve
        f1_path = results_path / 'F1_curve.png'
        if f1_path.exists():
            img = Image.open(f1_path)
            axes[1, 1].imshow(img)
            axes[1, 1].set_title('F1 Score Curve', fontsize=14, fontweight='bold')
            axes[1, 1].axis('off')
        
        plt.tight_layout()
        plt.savefig('results/yolo_training_results.png', dpi=300, bbox_inches='tight')
        print("‚úÖ Training results visualization saved!")
        plt.show()


# Example usage
if __name__ == "__main__":
    # Define class names for defects
    class_names = ['Dust', 'Bird-Drop', 'Physical-Damage', 'Electrical-Damage']
    
    # Initialize detector
    detector = SolarPanelObjectDetector(
        data_yaml_path='data/yolo_dataset/data.yaml',
        model_size='n'  # Use nano model for faster training
    )
    
    # Create data.yaml
    detector.create_data_yaml(
        dataset_path='data/yolo_dataset',
        class_names=class_names
    )
    
    # Prepare dataset structure
    detector.prepare_yolo_dataset(
        source_dir='data/processed',
        output_dir='data/yolo_dataset'
    )
    
    print("\n" + "="*70)
    print("‚ö†Ô∏è  NEXT STEPS:")
    print("="*70)
    print("1. Annotate your images using LabelImg or Roboflow")
    print("2. Save annotations in YOLO format (.txt files)")
    print("3. Place images in: data/yolo_dataset/images/{train,val,test}")
    print("4. Place labels in: data/yolo_dataset/labels/{train,val,test}")
    print("5. Then uncomment and run the training code below")
    print("="*70)
    
    # Uncomment below after annotation:
    """
    # Load pretrained model
    detector.load_model(pretrained=True)
    
    # Train
    results = detector.train(epochs=100, imgsz=640, batch=16)
    
    # Validate
    metrics = detector.validate()
    
    # Make predictions
    detector.predict('path/to/test/image.jpg', conf=0.25)
    
    # Visualize
    detector.visualize_predictions('path/to/test/image.jpg', save_path='results/detection.png')
    
    # Plot results
    detector.plot_training_results()
    
    # Export model
    detector.export_model(format='onnx')
    """
```
